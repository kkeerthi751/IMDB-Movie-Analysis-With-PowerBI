```python
# looking for the amount of column-wise null data in the pandas data
movies_data.isnull().sum().sort_values(ascending=False)
```

```python
percent_missing = movies_data.isnull().sum() * 100 / len(movies_data)
missing_values = pd.DataFrame({'percent_missing': percent_missing})
missing_values.sort_values(by ='percent_missing' , ascending=False)
```

```python
# Assuming 'data' is your DataFrame
selected_columns = ['title_year', 'num_user_for_reviews']

# Drop rows where either 'title_year' or 'num_user_for_reviews' is null
movies_data.dropna(subset=selected_columns, inplace=True)
```

```python
movies_data['title_year']
```

```python
# correcting data type of columns
movies_data['title_year'] =movies_data['title_year'].astype("int64")
```

```python
movies_data.dtypes
```

```python
movies_data.columns # show all columns name
```

```python
numerical_columns = ['num_critic_for_reviews', 'duration', 'director_facebook_likes', 'actor_3_facebook_likes',
                     'actor_1_facebook_likes', 'gross', 'facenumber_in_poster', 'budget',
                     'actor_2_facebook_likes', 'aspect_ratio']

for column in numerical_columns:
    # Check if the column is numeric before filling missing values
    if pd.api.types.is_numeric_dtype(movies_data[column]):
        movies_data[column].fillna(movies_data[column].mean(), inplace=True)
    else:
        # Handle non-numeric columns differently if needed
        print(f"Skipping non-numeric column: {column}")
```

```python
movies_data.isnull().sum()
```

```python
movies_data['director_name'].fillna('Unknown', inplace=True)
movies_data['actor_2_name'].fillna('Unknown', inplace=True)
movies_data['actor_1_name'].fillna('Unknown', inplace=True)
movies_data['actor_3_name'].fillna('Unknown', inplace=True)
```

```python
categorical_columns = ['color', 'plot_keywords', 'language', 'country', 'content_rating']
for column in categorical_columns:
   movies_data[column].fillna(movies_data[column].mode()[0], inplace=True)
```

```python
movies_data.isnull().sum()
```

```python
movies_data.drop_duplicates()
```

```python
movies_data.columns
```

```python
movies_data['profit'] = movies_data['gross'] - movies_data['budget']
```

```python
movies_data.drop(columns=['movie_imdb_link'], inplace=True)
```

```python
movies_data
```

```python
movies_data['genres'] = movies_data['genres'].str.split('|', expand=False).apply(lambda x: x[0] if x is not None else None)
```

```python
movies_data['plot_keywords'] = movies_data['plot_keywords'].str.split('|').apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)
```

```python
movies_data['movie_title'] = movies_data['movie_title'].str.split(':').apply(lambda x: x[0].strip() if x is not None else None)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'data'
# Replace 'data' with your actual DataFrame name

# Specify the integer columns
int_columns = ['num_voted_users', 'cast_total_facebook_likes', 'movie_facebook_likes']

# Set up subplots
fig, axes = plt.subplots(nrows=len(int_columns), ncols=1, figsize=(10, 6 * len(int_columns)))

# Plot distribution for each integer column
for i, column in enumerate(int_columns):
    sns.histplot(movies_data[column], kde=True, ax=axes[i], color='skyblue', bins=30)

    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
